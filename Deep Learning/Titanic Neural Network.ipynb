```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Load the Titanic dataset
train = pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv')

# Extract the target variable
y_train = train['Survived']
train.drop(labels='Survived', axis=1, inplace=True)

# Extract relevant features
columns_to_drop = ["Name", "Age", "SibSp", "Ticket", "Cabin", "Parch", "Embarked"]
train_test = train.append(test)
train_test.drop(labels=columns_to_drop, axis=1, inplace=True)

# One-hot encode 'Sex'
train_test_dummies = pd.get_dummies(train_test, columns=['Sex'])

# Fill missing values with 0.0
train_test_dummies.fillna(value=0.0, inplace=True)

# Separate back into training and test sets
x_train = train_test_dummies.values[0:891]
x_test = train_test_dummies.values[891:]

# Scale the features
scaler = MinMaxScaler()
x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.fit_transform(x_test)

# Split the training set into training and validation sets
x_train_sub, x_validation_sub, y_train_sub, y_validation_sub = train_test_split(
    x_train_scale, y_train, random_state=0
)

# Build a simple neural network model
model = Sequential()
model.add(Dense(64, input_dim=x_train_scale.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])

# Train the model
model.fit(x_train_sub, y_train_sub, epochs=20, batch_size=32, validation_data=(x_validation_sub, y_validation_sub))

# Evaluate the model on the validation set
_, accuracy = model.evaluate(x_validation_sub, y_validation_sub)
print(f'Accuracy score (validation): {accuracy:.3f}')
```
