```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
rng=np.random.RandomState(1)
x=np.dot(rng.rand(2,2),rng.randn(2,200)).T
plt.scatter(x[:,0],x[:,1])
plt.axis('equal')
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(x)
```

```python
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=2,
                    shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# plot data
plt.scatter(x[:, 0], x[:, 1], alpha=0.2)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal');
```

```python
print(pca.components_)
```

```python
pca = PCA(n_components=1)
pca.fit(x)
x_pca = pca.transform(x)
print("original shape:   ", x.shape)
print("transformed shape:", x_pca.shape)
```

```python
x_new = pca.inverse_transform(x_pca)
plt.scatter(x[:, 0], x[:, 1], alpha=0.2)
plt.scatter(x_new[:, 0], x_new[:, 1], alpha=0.8)
plt.axis('equal');
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

x, y = load_iris(return_X_y=True)
clf = LogisticRegression(random_state=0).fit(x, y)
clf.score(x, y)
```

```python
pca = PCA(n_components=0.95, whiten=True)
x_pca = pca.fit_transform(x)
```

```python
clf = LogisticRegression(random_state=0).fit(x_pca, y)
clf.score(x_pca, y)
```

```python
from sklearn.datasets import load_digits
digits=datasets.load_digits()
```

```python
digits.data.shape
```

```python
pca = PCA(2)  # project from 64 to 2 dimensions
projected = pca.fit_transform(digits.data)
print(digits.data.shape)
print(projected.shape)
```

```python
plt.scatter(projected[:, 0], projected[:, 1],
            c=digits.target, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('Spectral', 10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

```python
pca = PCA().fit(digits.data)
print(pca.explained_variance_ratio_)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
X, y = load_digits(return_X_y=True)
clf = LogisticRegression(random_state=0, max_iter=1000).fit(X, y)
clf.score(X, y)
```

```python
pca = PCA(20)  # project from 64 to 2 dimensions
pca_fit = pca.fit(digits.data)
projected = pca_fit.transform(digits.data)
print(projected.shape)
```

```python
clf = LogisticRegression(random_state=0, max_iter=1000).fit(projected, y)
clf.score(projected, y)
```
