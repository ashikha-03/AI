```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/Iris.csv')
```

```python
features=['SepalLengthCm', 'PetalLengthCm']
X=df[features]
```

```python
X_standardized = StandardScaler().fit_transform(X)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X_standardized)
clusters
```

```python
centroids = pd.DataFrame(X_standardized).groupby(clusters).mean().values
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X['SepalLengthCm'], X['PetalLengthCm'], c=clusters, cmap='viridis', edgecolors='k', s=50)
plt.title('DBSCAN Clustering of Iris Data')
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.show()
```
