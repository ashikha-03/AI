```python
X= dataset.iloc[:, [2,4]].values
X
```

```python
from sklearn.cluster import KMeans
wcss=[]

for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('Iris dataset')
plt.xlabel('Clusters')
plt.ylabel('Wcss')
plt.show()
```
